<b>Rewards calculation process, epsilon defines the exploration factor</b>

```python
def process_rewards(steps, bandits, levers, testbed, epsilon):
    initial_reward_estimates =  np.random.normal(testbed, 1)
    reward_estimates = np.zeros((bandits, levers))
    action_count = np.ones((bandits, levers)) 
    running_reward_sum = np.copy(reward_estimates) 
    
    optimal_choice_per_step = []
    optimal_choice_per_step.append(0)
    optimal_choice_per_step.append(0)

    mean_reward = 0

    Q_A = []
    Q_A.append(0) 
    Q_A.append(np.mean(initial_reward_estimates)) 

    for n in range(2, steps):
        
        sum_of_optimal_choice = 0 
        reward_sum_over_all_bandits_per_step = 0
        
        for problem_index in range(bandits):  
            
            if np.random.random() > epsilon:
                maxval = np.amax(reward_estimates[problem_index])
                maxval_indices = np.ravel(np.array(np.where(reward_estimates[problem_index] == maxval)))
                random_choice = np.random.choice( maxval_indices )
            else :
                 random_choice = np.random.randint(levers)

            # track optimal choice
            if random_choice == np.argmax(testbed[problem_index]):
                sum_of_optimal_choice += 1 
            
            running_reward_sum[problem_index][random_choice] += np.random.normal(testbed[problem_index][random_choice],1)
            action_count[problem_index][random_choice] += 1         
            Q = running_reward_sum[problem_index][random_choice] / action_count[problem_index][random_choice]
            reward_estimates[problem_index][random_choice] = Q

            reward_sum_over_all_bandits_per_step += Q

        Q_A.append((reward_sum_over_all_bandits_per_step) / bandits)
        optimal_choice_per_step.append((sum_of_optimal_choice / bandits)*100)

    return Q_A, optimal_choice_per_step
```

```python
np.random.seed(10)
bandits = 2000
levers = 10
steps = 2000
```

<b>Generate the rewards for different values of epsilons</b>

```python
epsilons= [0.0, 0.01, 0.1, 0.2, 0.5, 0.7, 0.9]
colors = ['r', 'g', 'b', 'k', 'y', 'c', 'lime']

fig, axs = plt.subplots(2, figsize=(15,15))

axs[0].set(xlabel="Steps", ylabel="Average Reward")
axs[1].set(xlabel="Steps", ylabel="% Optmial Action")

testbed = np.random.normal(0,1,(bandits, levers))
for index,eps in enumerate(epsilons):
    Q_A, optimal_choice_per_step = process_rewards(steps, bandits, levers, testbed, eps)
    axs[0].plot(np.arange(steps),Q_A, label=eps, color=colors[index])
    axs[1].plot(np.arange(steps),optimal_choice_per_step,label=eps, color=colors[index])

leg1 = axs[0].legend(loc="upper left", bbox_to_anchor=[0, 1],
                 ncol=2, shadow=True, title="Epsilon", fancybox=True)

leg2 = axs[1].legend(loc="upper left", bbox_to_anchor=[0, 1],
                 ncol=2, shadow=True, title="Epsilon", fancybox=True)
plt.show()
```
